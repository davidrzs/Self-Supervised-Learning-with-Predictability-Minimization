```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# cifar 10

X = [92.35, 92.70,92.03,91.24,91.04,91.75,90.80] #cl-lin-pred-missing
y = [0.503,0.457,0.444,0.340,0.258,0.599,0.484]
```

```python
# plot x,y and plot the linear best fit line
plt.scatter(X, y)
plt.xlabel('Performance')
plt.ylabel('Prediction')
plt.title('Performance vs Prediction')
plt.plot(np.unique(X), np.poly1d(np.polyfit(X, y, 1))(np.unique(X)), color='red')
plt.show()
```
