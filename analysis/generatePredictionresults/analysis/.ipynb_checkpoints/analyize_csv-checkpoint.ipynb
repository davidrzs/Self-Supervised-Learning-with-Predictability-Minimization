```python
def correlation_sum(corr_mat):
    assert corr_mat.shape[0] == corr_mat.shape[1]
    corr_mat_abs = np.absolute(corr_mat)
    sums = np.sum(corr_mat_abs, axis=1) - corr_mat_abs.diagonal() # axis doen't matter
    sums /= (corr_mat.shape[0]-1)
    return sums
```
